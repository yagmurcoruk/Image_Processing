```python
# görüntünün negatifini alan pyhton kodunu yazınız :
#görüntünün log dönüşüm kodunu yazınız :
 
import cv2
import numpy as np
import matplotlib.pyplot as plt

def negatifi(foto):
    L = np.max(foto)
    negatif_hal = L-foto
    return negatif_hal

def log(r,c):
    r = r.astype(float)
    s = c*np.log(1+r)
    return s.astype(np.uint8)

image = cv2.imread("brain.jpg",0)
image2 = cv2.imread("isik.jpg",0)

negatif_foto = negatifi(image)
log_donusumlu = log(image2,50)

yan_yana2 = np.hstack((image2,log_donusumlu))
yan_yana = np.hstack((image,negatif_foto))

#plt.imshow(yan_yana, cmap="gray")
plt.imshow(yan_yana2, cmap="gray")
```
