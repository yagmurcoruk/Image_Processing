```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def fotograf_negatifi(foto):
    L = np.max(foto)
    negatif = L-foto
    return negatif

foto = cv2.imread("mamo.jpg",0)  # ,0 eklendiğinde foto siyah beyaz olur.

print(foto.shape)   # fotoğraf rgb mi yoksa siyah beyaz mı baktık 

negatif = fotograf_negatifi(foto)  #fonksiyonu çağırdık

yan_yana = np.hstack((foto,negatif))  # yan yana göstermek için 

#plt.imshow(negatif, cmap="gray")
plt.imshow(yan_yana,cmap="gray")


cv2.waitKey(0)
cv2.destroyAllWindows()
```
